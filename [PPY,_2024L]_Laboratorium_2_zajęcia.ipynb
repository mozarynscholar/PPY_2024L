## 1.1. Zmienne i typy danych

W Pythonie zmienna to nazwa, która odnosi się do wartości. Python jest językiem dynamicznie typowanym, co oznacza, że nie trzeba deklarować typu zmiennej przed jej użyciem. Python wspiera różne typy danych,m.in.

* liczby całkowite (`int`),
* liczby zmiennoprzecinkowe (`float`),
* zmienne logiczne (`boolean`),
* łańcuchy znaków (`str`),
* listy (`list`),
* słowniki (`dict`)
* ...i wiele innych.

```python
liczba = 10                 # Typ int
zmiennoprzecinkowa = 20.5   # Typ float
napis = "Witaj świecie!"    # Typ str
lista = [1, 2, 3]           # Typ list
jestPiatek = True           # Typ boolean

print(liczba)
print(zmiennoprzecinkowa)
print(napis)
print(lista)
print(jestPiatek)
```

Czasem może być przydatne zadeklarowanie pustej zmiennej danego typu.

```python
pusty_napis = ""
print("Napis:"+pusty_napis)
```

* Do zadeklarowania pustej listy użyj pustych nawiasów kwadratowych:

```python
pusta_lista = []
print("Lista:", pusta_lista)
```

## 1.2. Instrukcje warunkowe

Instrukcje warunkowe pozwalają na wykonanie różnych sekcji kodu w zależności od spełnienia określonych warunków. Python używa standardowych słów kluczowych takich jak `if`, `elif` i `else`.

```python
# Przykład instrukcji warunkowej
liczba = 10

if liczba > 0:
    print("Liczba jest dodatnia")
elif liczba < 0:
    print("Liczba jest ujemna")
else:
    print("Liczba jest równa 0")
```

## 1.3. Pętle


Python obsługuje dwa główne typy pętli: `for` i `while`.

Pętle for są używane do iteracji po elementach sekwencji (takich jak listy, krotki, łańcuchy znaków) lub innych obiektów iterowalnych.

Pętla `while` wykonuje blok kodu, tak długo, jak spełniony jest określony warunek.

```python
# Przykład pętli for
print("pętla for:")
for i in range(5):  # Iteruje od 0 do 4
    print(i)

# Przykład pętli while
print("pętla while")
liczba = 5
while liczba > 0:
    print(liczba)
    liczba -= 1  # Dekrementacja liczby o 1
```

Do wprowadzaia danych z klawiatury służy funkcja `input()`, która czeka, aż użytkownik wpisze tekst na klawiaturze i naciśnie klawisz ENTER.

## 1.6. Komentarze

 Python obsługuje dwa główne typy komentarzy: komentarze jednoliniowe i wieloliniowe.

 **Komentarze jednoliniowe**

 Komentarze jednoliniowe rozpoczynają się od znaku hash (`#`). Wszystko, co znajduje się po tym znaku w danej linii, jest traktowane jako komentarz i nie jest wykonywane przez interpreter Pythona. Są one często używane do krótkich opisów kodu lub do tymczasowego wyłączania określonych linii kodu.

**Komentarze wieloliniowe**

 W Pythonie nie ma dedykowanego sposobu na tworzenie komentarzy wieloliniowych w taki sposób, jak w niektórych innych językach programowania, gdzie mogą one być otoczone specjalnymi znakami początkowymi i końcowymi. Zamiast tego, wieloliniowe „komentarze” są tworzone przy użyciu potrójnych cudzysłowów (`"""` lub `'''`), które technicznie są ciągami znaków wieloliniowymi, ale mogą być używane jako komentarze, jeśli nie są przypisane do żadnej zmiennej ani nie są używane w kodzie. Interpreter ignoruje takie nieprzypisane ciągi znaków, traktując je efektywnie jako komentarze.

 **Docstrings**

 Warto wspomnieć o specjalnym przypadku „komentarzy”, znanym jako docstrings (łańcuchy dokumentacyjne), które są technicznie ciągami znaków, ale służą do dokumentowania modułów, funkcji, klas i metod. Docstrings umieszcza się zaraz po definicji funkcji/modułu/klasy, używając potrójnych cudzysłowów. Docstrings mogą być następnie dostępne w czasie wykonywania programu za pomocą funkcji `help()` lub specjalnej zmiennej `__doc__`.

```python
# To jest komentarz jednoliniowy

print("Hello, world!")  # To także jest komentarz, znajdujący się w tej samej linii co kod

"""
To jest przykład komentarza wieloliniowego
służącego do opisu fragmentu kodu.
Możesz tutaj umieścić dłuższe wyjaśnienia.
"""
# jak odnieść się do tegok komentarza? czy w komentarzu może też być kod?
def funkcja():
    print("cokolwiek")
    """
    To jest komentarz wieloliniowy używany jako docstring.
    Opisuje, co robi funkcja.
    """
    pass

print(funkcja.__doc__)
```

## 2.1. Typy danych

### 2.1.1 Typy liczbowe


Numeryczne typy danych w Pythonie są używane do reprezentowania wartości liczbowych. Python udostępnia trzy podstawowe liczbowe typy danych - całkowite (`int`), zmiennoprzecinkowe (`float`) i urojone (`complex`). Te liczbowe typy danych umożliwiają wykonywanie różnych operacji arytmetycznych, takich jak dodawanie, odejmowanie, mnożenie i dzielenie. Zapewniają one niezbędną elastyczność do pracy z danymi numerycznymi w programach Python.

* `int` - przechowuje wartości całkowite, które mogą być dodatnie lub ujemne i nie zawierają przecinka dziesiętnego. Przykład: `num1 = 10`
* `float` - są to zmiennoprzecinkowe liczby rzeczywiste, które przechowują wartości dziesiętne. Składają się z części całkowitej i ułamkowej. Przykład: `fnum = 25.4`
* `complex` - są to liczby zespolone określone jako część rzeczywista i część urojona. Są one przechowywane w postaci `a+bj`, gdzie `a` jest częścią rzeczywistą, a `bj` reprezentuje część urojoną. Przykład: `num3= 2 + 3j`

**Funkcja `type()`**

Funkcja `type()` w Pythonie jest wbudowaną funkcją, która zwraca typ obiektu podanego jako argument. Typ obiektu w Pythonie określa, jakie operacje można na nim wykonać oraz jakie atrybuty posiada.

Funkcja `type()` może być używana zarówno do sprawdzenia typu obiektu, jak i do dynamicznego tworzenia nowych typów (klas).

Gdy używana jest do sprawdzenia typu, `type()` przyjmuje jeden argument i zwraca obiekt typu, który reprezentuje typ argumentu.

```python
num1 = 10
print(num1)
print("Dtype of num1 is",type(num1))

fnum = 26.5
print(fnum)
print("Dtype of fnum is",type(fnum))

num3 = 2 + 6j
print(num3)
print('Dtype of num3 is',type(num3))
```

### 2.1.2 Typy do reprezentowania łańcuchów znaków

Typ `string` w Pythonie jest jednym z wbudowanych typów danych służących do reprezentowania tekstowych danych, czyli ciągów znaków. W Pythonie, łąńcuchy znaków są niezmiennymi sekwencjami znaków, co oznacza, że raz utworzony ciąg znaków nie może być zmodyfikowany. Zamiast tego, każda operacja, która wydaje się modyfikować łańcuch znakw, tak naprawdę tworzy nowy łańcuch z wynikowymi zmianami.

Łańcuchy znaków w Pythonie można definiować na różne sposoby, używając pojedynczych ('), podwójnych (") lub potrójnych (''' lub """) cudzysłowów, co pozwala na łatwe tworzenie ciągów zawierających znaki cudzysłowów lub apostrofy bez konieczności wstawiania znaków specjalnych (escapowania), a także definiowanie wieloliniowych stringów.

Łańcuchy znaków  w Pythonie wspierają szeroki zakres operacji, takich jak konkatenacja (łączenie stringów), powielanie, dostęp do poszczególnych znaków przez indeksowanie, wycinanie fragmentów łańcucha (slicing), a także wiele metod umożliwiających przetwarzanie ciągów znaków, takich jak l`ower(), upper(), strip(), split(), join()` i wiele innych.

**Funkcja `len()`**

Funkcja `len()` w Pythonie jest używana do zwracania liczby elementów (długości) w obiekcie kolekcji, takim jak lista, krotka, słownik, zbiór, czy ciąg znaków (string).

```python
# Definicja łańcucha znaków
powitanie = "Hello, world!"

# Dostęp do znaku (indeksowanie)
pierwszy_znak = powitanie[0]  # 'H'

# Wycinanie fragmentu stringa
czesc = powitanie[0:5]  # 'Hello'

# Metody stringów
powitanie_malymi_literami = powitanie.lower()  # 'hello, world!'

print(pierwszy_znak)
print(czesc)
print(powitanie_malymi_literami)
print(len(powitanie))
```

### 2.1.3. Typy danych zbiorowych

W Pythonie do typów danych zbiorowych zaliczamy:

* **Listy (Lists)**: Zmienne sekwencje, które mogą przechowywać elementy różnych typów. Listy są zamówione, co oznacza, że zachowują kolejność dodawania elementów.
* **Krotki (Tuples)**: Niezmienne sekwencje, podobnie jak listy, ale nie można ich modyfikować po utworzeniu. Krotki również zachowują kolejność elementów.
* **Słowniki (Dictionaries)**: Zbiory par klucz-wartość, gdzie każdy klucz jest unikalny.
* **Zbiory (Sets)**: Nieuporządkowane kolekcje unikalnych elementów. Zbiory są używane do przechowywania i operowania na zbiorach elementów, umożliwiają wykonanie operacji takich jak unia, przecięcie, różnica.

#### 2.1.3.1 Listy

Typ `list` w Pythonie jest jednym z wbudowanych typów danych zbiorowych (kolekcyjnych), który służy do przechowywania wielu elementów jako pojedynczą jednostkę. Lista jest uporządkowaną kolekcją, co oznacza, że **elementy w liście są przechowywane w określonej kolejności, od lewej do prawej, począwszy od indeksu 0**. Listy są zmienne (mutable), co oznacza, że można modyfikować ich zawartość po utworzeniu, dodając, usuwając lub zmieniając elementy.

**Kluczowe cechy list w Pythonie:**

* **Uporządkowane:** Kolejność elementów w liście ma znaczenie i jest zachowana.
* **Zmienne:** Można zmieniać, dodawać i usuwać elementy w liście.
* **Elastyczne:** Listy mogą przechowywać elementy różnych typów, w tym inne listy, co umożliwia tworzenie list zagnieżdżonych.
* **Dynamiczne**: Rozmiar listy może być zmieniany dynamicznie, co oznacza, że Python automatycznie zarządza alokacją pamięci dla list.

**Tworzenie listy:**

Listę można utworzyć poprzez umieszczenie elementów w nawiasach kwadratowych `[]`, rozdzielając je przecinkami.

```python
moja_lista = [1, 2, 3, 4, 5]
print(moja_lista)

lista_str = ['apple', 'banana', 'cherry']
print(lista_str)

lista_zag = [1, 'apple', [3.14, 42, ['a', 'b', 'c']], 'dog']
print(lista_zag)
type(lista_zag)
```

Operacje na listach:

Listy oferują szereg metod umożliwiających manipulowanie ich zawartością, w tym:

* Dodawanie elementów: `append(), insert(), extend()`
* Usuwanie elementów: `remove(), pop(), del`
* Sortowanie: `sort()`
* Odwracanie: `reverse()`
* Wyszukiwanie: `index(), in`
* Liczenie wystąpień: `count()`

```python
lista = [1, 2, 3]
lista.append(4)  # Dodaje 4 na końcu listy
print(lista)  # [1, 2, 3, 4]

lista.insert(1, 'a')  # Wstawia 'a' na pozycji o indeksie 1
print(lista)  # [1, 'a', 2, 3, 4]

lista.remove('a')  # Usuwa pierwsze wystąpienie 'a'
print(lista)  # [1, 2, 3, 4]

element = lista.pop(2)  # Usuwa i zwraca element na pozycji o indeksie 2
print(element)  # 3
print(lista)  # [1, 2, 4]
```

### 2.1.4. Typ logiczny





W Pythonie typ logiczny jest reprezentowany przez typ `bool`, który przyjmuje jedną z dwóch wartości: `True` (prawda) lub `False` (fałsz). Te wartości są wykorzystywane do reprezentowania prawdy lub fałszu w operacjach logicznych i warunkowych.

Typ `bool` jest podtypem liczby całkowitej (`int`), gdzie `True` jest traktowane jak `1`, a `False` jak `0`. Oznacza to, że wartości bool mogą być używane w operacjach arytmetycznych i porównaniach.

```python
# Przykładowe wartości logiczne
prawda = True
fałsz = False

# Użycie w instrukcji warunkowej
if prawda:
    print("To jest prawda.")
else:
    print("To jest fałsz.")

# Operacje logiczne
a = 5
b = 10

# Sprawdzenie równości
czy_rowne = (a == b)  # Zwróci False, ponieważ 5 nie jest równe 10

# Sprawdzenie większości
czy_a_wieksze_od_b = (a > b)  # Zwróci False, ponieważ 5 nie jest większe od 10
```

### 2.1.5. Konwersja typów

Konwersja typów w Pythonie odnosi się do zmiany typu danych jednego obiektu na inny. Jest to często konieczne, gdy operuj się na danych, które zostały wcześniej zdefiniowane jako typ inny niż ten, którego wymaga dana operacja lub algorytm. Python oferuje wbudowane funkcje, które pozwalają na łatwą konwersję między podstawowymi typami danych, takimi jak liczby całkowite (`int`), liczby zmiennoprzecinkowe (`float`), ciągi znaków (`str`) i typy logiczne (`bool`).

**Funkcje służące do konwersji typów**

* `int()`, aby przekonwertować wartość na typ całkowity
* `float()`, aby przekonwertować wartość na typ zmiennoprzecinkowy
* `str()`, aby przekonwertować wartość na ciąg znaków
* `bool()`, aby przekonwertować wartość na typ logiczny

**Uwagi**

Wartości typu None, puste kolekcje (`[], {}, (), ''`), oraz liczba `0` są konwertowane na `False`, a wszystkie inne wartości na `True`.

Próba konwersji niektórych typów danych może spowodować błąd, jeśli konwersja nie ma sensu lub nie jest możliwa. Na przykład, próba konwersji ciągu znaków, który nie reprezentuje liczby (np. `"tekst"`) na typ liczbowy (`int` lub `float`) spowoduje `ValueError`.

Konwersja typów jest często używana podczas wczytywania danych z plików tekstowych lub sieci, gdzie wszystko jest domyślnie traktowane jako ciąg znaków, oraz podczas interakcji z użytkownikiem poprzez funkcję `input()`, która również zwraca dane wejściowe jako ciąg znaków.

```python
x = "10"
y = int(x)  # Konwersja ciągu znaków na liczbę całkowitą
print(y, type(y))

x = "3.14"
y = float(x)  # Konwersja ciągu znaków na liczbę zmiennoprzecinkową
print(y, type(y))

x = 10
y = str(x)  # Konwersja liczby całkowitej na ciąg znaków
print(y, type(y))

x = 0
y = bool(x)  # Konwersja liczby całkowitej na wartość logiczną (0 konwertuje się na False)
print(y, type(y))

x = 1
y = bool(x)  # Inne wartości niż 0 konwertują się na True
print(y, type(y))
```

Pozwalają wykonywać operacje matematyczne jak dodawanie, odejmowanie etc.

Kolejność operacji (zwana również pierwszeństwem) operatorów matematycznych Pythona jest podobna do tej w matematyce. Operator `**` jest obliczany jako pierwszy; operatory `*`, `/`, `//` i `%` są obliczane jako następne, od lewej do prawej; a operatory `+` i `-` są obliczane jako ostatnie (również od lewej do prawej). W razie potrzeby można użyć nawiasów, aby zastąpić zwykłe pierwszeństwo. Biała spacja pomiędzy operatorami i wartościami nie ma znaczenia dla Pythona (z wyjątkiem wcięcia na początku linii), ale pojedyncza spacja jest konwencją.

**Przykład:**

obraz.png

```python
a = 7
b = 2

# dodawanie
print ('Dodawanie: ', a + b)

# odejmowanie
print ('Odejmowanie: ', a - b)

# mnożenie
print ('Mnożenie: ', a * b)

# dzielenie
print ('Dzielenie: ', a / b)

# reszta z dzielenia
print ('Reszta z dzielenia:', a % b)

# a do potęgi b
print ('Podniesienie do potęgi: ', a ** b)

# dzielenie z zaokrągleniem w dół
print ('Dzielenie z zaokrągleniem w dół: ', a // b)

liczba=(5 - 1) * ((7 + 1) / (3 - 1))
print("wynik wyrażenia:", liczba)
```

### 2.2.3. Operatory relacyjne

Operatory relacyjne stosujemy w sytuacje, gdzie jest potrzeba porównania dwóch elementów. Najczęściej używane są w instrukcjach warunkowych i iteracyjnych.

# 3 Normy programowania w języku Python

**Normy dotyczącze programowania w języku Python znajdują się w tzw. PEP**

PEP w kontekście Pythona oznacza *Python Enhancement Proposal*, czyli Propozycję Ulepszenia Pythona. Jest to projekt dokumentu dostarczający informacji do społeczności Pythona, lub opisujący nową funkcjonalność bądź zmiany w języku Python. PEP-y są głównym mechanizmem do proponowania dużych nowych funkcji, do zbierania opinii społeczności na ich temat oraz dokumentowania decyzji projektowych, które zostały podjęte w Pythonie. Każdy PEP powinien dostarczyć jasny i konkretyny opis proponowanej funkcjonalności lub zmiany oraz uzasadnienie jej wprowadzenia.

**Są trzy główne typy PEP-ów:**

* PEP-y standardowe (*Standards Track PEPs*) dotyczące zmian w języku Python, standardowej bibliotece, lub procesów i środowiska tworzenia Pythona.
* PEP-y informacyjne (*Informational PEPs*) dla wytycznych lub informacji do społeczności Pythona, które nie wprowadzają zmian w samym Pythonie.
* PEP-y proceduralne (*Process PEPs*) opisujące zmiany w procedurach zarządzających projektem Python, lub służące do ustanowienia nowych procesów.

Proces PEP zaczyna się od nowego pomysłu lub propozycji ulepszenia Pythona. Autor PEP-u tworzy dokument według określonego szablonu, który jest następnie publikowany w repozytorium PEP-ów Pythona i przedstawiany społeczności oraz komitetowi sterującemu Pythona do rozpatrzenia. Po dyskusji i ewentualnych poprawkach, PEP może zostać zaakceptowany, odrzucony lub wycofany przez autora.
