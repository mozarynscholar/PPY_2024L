# 1.Obsługa wyjątków


Obsługa wyjątków w Pythonie jest mechanizmem, który umożliwia programowi radzenie sobie z nieoczekiwanymi błędami w trakcie jego wykonania. Pozwala to na kontrolowanie przepływu programu nawet w przypadku wystąpienia błędu i zapewnia większą odporność aplikacji.

Podstawowe składniki obsługi wyjątków:

* `try`: Blok kodu, który jest monitorowany pod kątem błędów. Jeśli Python napotka błąd w bloku try, przerwie jego wykonanie i przejdzie do bloku except.

* `except`: Blok kodu, który jest wykonywany, jeśli w bloku try pojawi się wyjątek. Można określić typ wyjątku, który chce się obsłużyć. Jeśli nie podano typu, domyślnie blok except złapie każdy wyjątek.

* `else`: Opcjonalny blok, który zostanie wykonany, jeśli w bloku try nie wystąpi żaden wyjątek.

* `finally`: Opcjonalny blok, który zostanie wykonany na końcu, niezależnie od tego, czy w bloku try wystąpił wyjątek, czy nie. Jest to miejsce na kod, który musi być wykonany, na przykład dla zwolnienia zasobów.

Przykład

```python
try:
    # próbujemy podzielić przez zero
    result = 10 / 0
except ZeroDivisionError:
    # obsługa dzielenia przez zero
    print("Nie można dzielić przez zero!")
else:
    # kod wykonuje się, gdy nie wystąpi wyjątek
    print("Dzielenie przebiegło pomyślnie!")
finally:
    # kod wykonuje się zawsze, niezależnie od wyniku
    print("To jest blok 'finally', który zawsze się wykonuje.")
```

**Wskazówki dotyczące obsługi wyjątków**

* **Specyfika wyjątków**: Zawsze należy być jak najbardziej specyficznym w deklaracji wyjątków, jakie chcemy przechwycić. Nalezy unikać używania ogólnego `except:,` który złapie każdy możliwy wyjątek, co może utrudnić debugowanie.

* **Logowanie**: Podczas obsługi wyjątków warto rejestrować błędy, na przykład używając modułu `logging`. Pozwoli to na lepsze zrozumienie kontekstu, w jakim wyjątek wystąpił.

* **Przemyślane rzucanie wyjątków**: W przypadku tworzenia własnych funkcji lub klas, możesz również używać słowa kluczowego `raise` do rzucania wyjątków, gdy funkcja napotka na stan, którego nie może obsłużyć.

* **Dokumentacja Pythona**: Najbardziej wiarygodnym źródłem informacji o wyjątkach w Pythonie jest oficjalna dokumentacja Pythona. Zawiera ona kompletną listę wbudowanych wyjątków wraz z opisem, kiedy są one wywoływane. Oficjalna dokumentacja wyjątków: [Python 3 Exceptions](https://docs.python.org/3/library/exceptions.html).

* **Moduł `builtins`**: Python przechowuje wszystkie wbudowane wyjątki jako klasy w module `builtins`. Można użyć funkcji `dir()` na module `builtins`, aby zobaczyć listę dostępnych wyjątków.

## 1.3. Rzucanie wyjątków

Słowo kluczowe `raise` w Pythonie pozwala na ręczne wywołanie wyjątków w kodzie. Można używać `raise`, aby zgłosić błąd, kiedy kod napotka na stan, który nie jest obsługiwany normalnie przez funkcję. Jest to szczególnie użyteczne w walidacji danych lub w sytuacjach, które wymagają przerwania wykonania programu i przekazania informacji o błędzie.

Możesz także dołączyć dodatkowe informacje do wyjątku, aby pomóc w debugowaniu.

**Przykład 1**

```python
def divide(x, y):
#Jeśli funkcja divide otrzyma y równe zero, wywoła wyjątek ZeroDivisionError z odpowiednią wiadomością.
    if y == 0:
        raise ZeroDivisionError("Nie można dzielić przez zero")
    return x / y
```

# 2.Iteratory i Generatory

## 2.1.Iteratory

**W Pythonie iterator to obiekt, który reprezentuje sekwencję wartości i pozwala na pobieranie ich po jednej w sposób sekwencyjny.**

Główne cechy iteratorów:

* Dostęp sekwencyjny: Iteratory umożliwiają pobieranie kolejnych elementów sekwencji jeden po drugim, bez konieczności przechowywania całej sekwencji w pamięci naraz.
* Oszczędność pamięci: Iteratory generują elementy w momencie ich użycia, co oznacza, że nie są one generowane wszystkie na raz. Oszczędza to pamięć, szczególnie w przypadku dużych sekwencji.
* Protokół iteratora: Iteratory definiują protokół, który określa dwie metody:
** `__iter__()`: Zwraca obiekt iteratora.
** `__next__()`: Zwraca kolejny element sekwencji i przesuwa wewnętrzny wskaźnik iteratora na następny element.

Przykład:

Listy w Pythonie są domyślnie iteratorami. Oznacza to, że można po nich iterować za pomocą pętli `for`:

```python
liczby = [1, 2, 3, 4, 5]

for liczba in liczby:
  print(liczba)
```

## 2.2.Generatory

Generatory w Pythonie to **specjalny rodzaj funkcji**, które pozwalają na **generowanie sekwencji danych**, ale w sposób leniwy (*lazy evaluation*). Oznacza to, że generatory produkują wartości jeden po drugim, w miarę potrzeby, zamiast generować wszystkie wartości naraz. Dzięki temu są one szczególnie użyteczne przy obsłudze dużych zbiorów danych lub przy realizacji operacji, które wymagają sekwencyjnego przetwarzania danych bez konieczności przechowywania wszystkich elementów w pamięci.

Generator jest obiektem, który można stworzyć na dwa sposoby:

* wykonując funkcję generującą,
* wykonując wyrażenie generujące.

Funkcja, która definiuje generator, używa słowa kluczowego `yield` zamiast `return`. Słowo `yield` przekazuje wartość z funkcji, ale w przeciwieństwie do `return`, nie kończy działania funkcji. Funkcja zostaje zawieszona i może być wznowiona później, aby kontynuować generowanie wartości od miejsca, w którym została zawieszona.

**Przykład**

```python
def generuj_kwadraty(n):
  for i in range(n):
    yield i * i

# Utworzenie generatora (obiektt)
generator = generuj_kwadraty(5)

# Pobieranie kolejnych elementów sekwencji
print(next(generator))  # 0
print(next(generator))  # 1
print(next(generator))  # 4
print(next(generator))  # 9
print(next(generator))  # 16
print(next(generator))  #wyjątek
```

W podanym przykładzie funkcja `next()` służy do pobierania kolejnych elementów sekwencji generowanej przez generator generuj_kwadraty. Generator ten zwraca obiekt iteratora, który umożliwia iterowanie po sekwencji kwadratów liczb od 0 do n-1, gdzie n jest argumentem przekazanym do funkcji generatora.

Funkcja `next()` pobiera kolejny element z sekwencji za każdym razem, gdy jest wywoływana. Dzieje się tak ponieważ:

* **Generator pamięta swój stan:** Kiedy funkcja generatora jest wywoływana po raz pierwszy, wykonuje się kod do momentu napotkania instrukcji yield. W tym momencie wartość `i * i` (czyli kwadrat bieżącej wartości `i`) jest zwracana do funkcji `next()`, a wykonanie funkcji generatora jest zawieszone.
* **Przy kolejnych wywołaniach**: Kiedy funkcja `next()` jest wywoływana ponownie, **wykonanie funkcji generatora jest wznowane od miejsca, w którym zostało zawieszone**. Następnie i jest zwiększane o 1, kod jest wykonywany dalej do następnej instrukcji `yield`, a nowa wartość `i * i` jest zwracana do funkcji `next()`.

W ten sposób funkcja `next()` pozwala na iterowanie po sekwencji generowanej przez generator bez konieczności generowania całej sekwencji naraz. Jest to szczególnie przydatne w przypadku dużych sekwencji, ponieważ oszczędza pamięć i czas.

**Należy pamiętać, że po pobraniu wszystkich elementów sekwencji (w tym przypadku 5 elementów), kolejne wywołanie `next(generator)` spowoduje wyjątek `StopIteration`, ponieważ w sekwencji nie ma już więcej elementów.**

**Przykład 2**

Aby użyć generatora, najczęściej korzysta się z pętli `fo`r, która automatycznie pobiera kolejne wartości z generatora

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

for number in count_up_to(5):
    print(number)
```

## 2.3. Różnice pomiędzy iteratorami a generatorami

**Iterator:**

* **Iterator to obiekt**, który implementuje metody `__iter__()` i `__next__()`. Metoda `__iter__()` zwraca iterator, a metoda `__next__()` zwraca kolejne elementy sekwencji. Gdy nie ma więcej elementów, `__next__()` wyrzuca wyjątek `StopIteration`.
* Może być zaimplementowany jako klasa z odpowiednimi metodami.
* Są bardziej elastyczne w zastosowaniach, gdyż można je projektować jako bardziej złożone struktury danych z dodatkowymi funkcjonalnościami.
* Może być zarówno leniwie wykonywany (*lazy evaluation*), jak i w pełni zainicjowany w pamięci (zależnie od implementacji).

**Generator:**

* **Generator to funkcja**, która używa słowa kluczowego `yield` do przekazywania wartości. Generatory automatycznie implementują metody `__iter__()` i `__next__()`, więc **nie trzeba ich jawnie definiować**.
* Jest zazwyczaj prostszy w implementacji niż klasa iteratora, ponieważ wymaga mniejszej ilości kodu i jest bardziej bezpośredni w użyciu.
* Zawsze wykonuje leniwie (*lazy evaluation*), co oznacza, że wartości są generowane na żądanie. To sprawia, że generatory są bardzo efektywne pod względem użycia pamięci, szczególnie przy dużych lub nieskończonych sekwencjach danych.
* Idealne do prostych zadań wymagających sekwencyjnego generowania danych, bez potrzeby przechowywania wszystkich danych w pamięci jednocześnie.

# 3.Dekoratory

Dekoratory w Pythonie to narzędzie, które pozwala modyfikować zachowanie funkcji, metod i klas, dodając dodatkową funkcjonalność przed i po ich wywołaniu, bez zmiany samej definicji funkcji. **Dekoratory są wyrażeniem wyższego rzędu, co oznacza, że akceptują funkcję jako argument i zwracają nową funkcję.** Pozwala to na rozszerzanie i modyfikowanie zachowań funkcji w elastyczny i ponownie używalny sposób.

**Podstawy dekoratorów**

Dekorator to zasadniczo funkcja, która przyjmuje inną funkcję jako argument i zwraca jeszcze inną funkcję. Dekorator "owija" oryginalną funkcję, co pozwala na wykonanie dodatkowego kodu przed i po oryginalnej funkcji, nie modyfikując jej bezpośrednio.

**Przykład**

W tym przykładzie, `say_hello` jest "owinięta" przez `wrapper` funkcję zdefiniowaną wewnątrz `my_decorator`. Kiedy wywołujemy `say_hello`, wywołujemy `wrapper`, który wykonuje kod przed i po wywołaniu oryginalnej funkcji `say_hello`.

```python
def my_decorator(func):
#argumentem jest fukncja
    def wrapper():
    #funkcja wewnętrzna, "owijająca"
        print("Coś się dzieje przed wywołaniem funkcji.")
        func() #wywołanie funkcji z argumentu
        print("Coś się dzieje po wywołaniu funkcji.")
    return wrapper

def say_hello():
    print("Hello!")

# Dekorowanie funkcji
say_hello = my_decorator(say_hello)

say_hello()
```

**Składnia dekoratorów**

Python oferuje specjalną składnię dla dekoratorów `@nazwa_dekoratora`, która ułatwia ich używanie. Powyższy przykład można przepisać używając składni dekoratora, co czyni kod bardziej zrozumiałym i klarownym.

**Przykład**

Użycie `@my_decorator` przed definicją funkcji `say_hello` jest równoważne z

`say_hello = my_decorator(say_hello) `

ale jest bardziej czytelne i bezpośrednie.

```python
def my_decorator(func):
#argumentem jest fukncja
    def wrapper():
    #funkcja wewnętrzna, "owijająca"
        print("Coś się dzieje przed wywołaniem funkcji.")
        func() #wywołanie funkcji z argumentu
        print("Coś się dzieje po wywołaniu funkcji.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

**Dekoratory z argumentami**

Czasami trzeba, aby dekorowana funkcja mogła przyjmować argumenty. Dekorator można dostosować, by obsługiwał argumenty

**Przykład**

W tym przypadku, `wrapper` używa `*args` i `**kwargs` do przechwytywania wszystkich `pozycyjnych i nazwanych` argumentów przekazywanych do funkcji `greet` i przekazuje je do oryginalnej funkcji.

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Coś się dzieje przed wywołaniem funkcji.")
        result = func(*args, **kwargs)
        print("Coś się dzieje po wywołaniu funkcji.")
        return result
    return wrapper

@my_decorator
def greet(name):
    print(f"Hello {name}!")

greet("Alice")
```

**Wiele dekoratorów**

Można również stosować wiele dekoratorów do jednej funkcji. Każdy dekorator modyfikuje zachowanie funkcji w określony sposób, a gdy stosujemy ich kilka, każdy kolejny dekorator "owija" wynik poprzedniego.

**Przykład**

Załóżmy, że mamy dwa dekoratory: jeden, który loguje argumenty funkcji, a drugi, który mierzy czas wykonania funkcji.

```python
import time

def log_args(func):
    def wrapper(*args, **kwargs):
        args_repr = [repr(a) for a in args]                        # 1
        kwargs_repr = [f"{k}={v!r}" for k, v in kwargs.items()]    # 2
        signature = ", ".join(args_repr + kwargs_repr)             # 3
        print(f"Calling {func.__name__}({signature})")
        result = func(*args, **kwargs)
        print(f"{func.__name__} returned {result!r}")              # 4
        return result
    return wrapper

def time_execution(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Executed {func.__name__} in {end_time - start_time:.4f}s")
        return result
    return wrapper

@log_args
@time_execution
def multiply(x, y):
    return x * y

multiply(5, 7)
```

**Różnica między `str()` a `repr()`**

* `str()` - Stosowane do **stworzenia "ładnej", czytelnej dla ludzi reprezentacji obiektu**. Zazwyczaj jest bardziej zwięzłe i skupia się na wyświetlaniu informacji w przyjaznej formie.
* `repr()` - Stosowane do generowania **oficjalnej reprezentacji obiektu, która może być często użyta do odtworzenia tego samego obiektu** (nie zawsze to możliwe, ale to częsty cel). Jest to bardziej techniczna i dokładna reprezentacja. Jeśli masz obiekt, który chcesz sformatować w ciąg znaków i potrzebujesz jego reprezentacji `repr`, możesz użyć `obiekt!r`.

**Wyjaśnienie**

1. Dekorator `log_args`:
* Loguje nazwę funkcji, jej argumenty pozycyjne i nazwane, a następnie wynik jej działania.
* Argumenty są konwertowane na reprezentacje stringowe (`repr`), co ułatwia ich czytanie w logu.

2. Dekorator `time_execution`:
* Mierzy czas, jaki upłynął od wywołania funkcji do zwrócenia wyniku.
* Loguje czas wykonania funkcji.

3. Kolejność dekoratorów:
* Dekoratory są stosowane od dołu do góry. Najpierw funkcja `multiply` jest owijana przez `time_execution`, a następnie wynikowa funkcja jest owijana przez `log_args`.
* Oznacza to, że `log_arg`s jest najbardziej zewnętrznym dekoratorem i jako pierwszy rejestruje wywołanie i wynik funkcji, a następnie time_execution mierzy czas wykonania tego, co zostało zarejestrowane przez `log_args`.

4. Efekt końcowy:
* Gdy wywołasz funkcję `multiply(5, 7)`, najpierw zalogowane zostaną argumenty i nazwa funkcji, następnie zostanie zmierzony czas wykonania, a na koniec zalogowany zostanie zwrócony wynik.
