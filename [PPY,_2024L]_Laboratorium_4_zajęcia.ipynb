# 1.Operacje wejścia-wyjścia (obsługa plików)

Operacje na plikach w Pythonie pozwalają na tworzenie, czytanie, aktualizowanie i usuwanie plików. Aby pracować z plikami, Python oferuje wbudowany mechanizm, pozwalający na podstawowe operacje na plikach


## 1.1. Otwieranie pliku


Do otwierania pliku się funkcji `open()`, która przyjmuje dwa główne argumenty: nazwę pliku i tryb otwarcia.

Składnia jest następująca

`file_object = open(file_name, mode)`

gdzie

* `file_name:` Ścieżka do pliku, który chcesz otworzyć. Może być to nazwa pliku w bieżącym katalogu, w którym jest uruchomiony skrypt Pythona, lub ścieżka do pliku w innym miejscu na dysku.
* `mode:` Tryb, w jakim plik ma być otwarty. Określa, czy plik ma być otwarty do odczytu, zapisu, czy do dodawania zawartości.

Tryby otwarcia to m.in.:

* `'r'` - do czytania (domyślny),
* `'w'` - do pisania (nadpisuje plik),
* `'a'` - do dopisywania (dodaje na końcu pliku),
* `'r+'` - do czytania i pisania w pliku.
* Dodając `b` do trybu (np. `rb`, `wb`, `ab`), plik zostanie otwarty w trybie binarnym, co jest przydatne przy pracy z plikami, które nie są tekstem, na przykład obrazami czy plikami dźwiękowymi.

## 1.2. Zamykanie pliku



Po zakończeniu pracy z plikiem należy go zamknąć za pomocą metody `close()`.

````python
file_path = '/content/drive/MyDrive/Colab Notebooks/test.txt'

plik = open(file_path, 'r')
zawartosc = plik.read()
plik.close() #zamknięcie pliku

print(zawartosc)
````

## 1.3. Automatyczne zarządzanie zasobami

Aby automatycznie zarządzać zasobami i upewnić się, że plik zostanie zamknięty, można użyć instrukcji `with`. Python automatycznie zamknie plik po wyjściu z bloku `with`.

Instrukcja `with` w Pythonie jest używana w kontekście zarządzania zasobami, oferując wygodny sposób na automatyczne zarządzanie otwieraniem i zamykaniem zasobów, takich jak **pliki**, **połączenia sieciowe** lub **blokady**. Jest to zaimplementowane przy użyciu tzw. **menedżerów kontekstu**, definiujących działania do wykonania przed wejściem do bloku kodu oraz po jego opuszczeniu, niezależnie od tego, czy opuszczenie nastąpiło z powodu zakończenia działania kodu, czy z powodu wystąpienia wyjątku.

````python
file_path = '/content/drive/MyDrive/Colab Notebooks/test.txt'

with open(file_path, 'r') as plik: # 'r' jest skrótem od słowa 'read'
    #po zakończeniu tego bloku, plik zostanie automatycznie zamknięty
    zawartosc = plik.read()

print(zawartosc)
````

## 1.4. Odczyt z pliku

Do czytania z pliku służą metody takie jak

* `read()` - odczytanie całej zawartości pliku
* `readline()` - odczytanie pojedynczej linii,

Aby zapisać coś do pliku, używa się metod

Jeśli wymagany jest koniec linii, należy dodać go samodzielnie (`\n`) do każdego ciągu w iterowalnym obiekcie.

## 1.7. Tryby specjalne

W Pythonie, tryby 'r+', 'w+', i 'a+' przy otwieraniu plików służą do obsługi plików w sposób umożliwiający jednoczesny odczyt i zapis. Główne różnice między tymi trybami dotyczą tego, czy plik jest czyszczony przy otwarciu, czy zostanie utworzony, jeśli nie istnieje, oraz gdzie jest umieszczony kursor dla operacji zapisu.

**'r+' - Tryb odczytu i zapisu**

* **Odczyt i zapis:** Ten tryb pozwala na odczytywanie z pliku oraz zapisywanie do niego.
* **Plik musi istnieć:** Próba otwarcia pliku, który nie istnieje, zakończy się błędem.
* **Pozycjonowanie na początku:** Kursor jest umieszczony na początku pliku, co oznacza, że zapis zacznie się od początku pliku i może nadpisać istniejącą treść, jeśli nie zostanie najpierw przesunięty.

**'w+' - Tryb zapisu i odczytu**

* **Zapis i odczyt:** Podobnie jak 'r+', pozwala na odczyt i zapis.
* **Tworzy plik, jeśli nie istnieje:** Jeśli plik nie istnieje, zostanie utworzony.
* **Czyszczenie zawartości:** Otwierając plik, 'w+' usuwa jego zawartość. Jeśli plik już istniał, jego poprzednia treść zostanie usunięta.

**'a+' - Tryb dopisywania i odczytu**

* **Dopisywanie i odczyt:** Umożliwia dopisanie treści na końcu pliku oraz odczytywanie z pliku.
* **Tworzy plik, jeśli nie istnieje:** Tak jak w przypadku 'w+', jeśli plik nie istnieje, zostanie utworzony.
* **Pozycjonowanie na końcu dla zapisu:** Kursor dla operacji zapisu jest umieszczony na końcu pliku, więc wszystkie zapisy dodają treść na jego końcu. Możliwy jest jednak odczyt z dowolnego miejsca w pliku.

## 2.1 Instrukcje warunkowe


Konstrukcja `if ... else` pozwala nam tworzyć instrukcje warunkowe - np. do rozwiązywania problemów gdzie należy podjąć jakąś decyzję.

W programowaniu korzystamy z polecenia `if` jeśli dany blok kodu ma zostać wywołany tylko w przypadku spełnienia warunku. Warunek ma wartość logiczną (pusta lista, krotka,... jest interpretowana jako False)

Ogóly format:

```
if warunek1:
    zagnieżdżony blok instrukcji sterującej
elif warunek2:
    zagnieżdżony blok kodu
elif warunek 3:
    zagnieżdżony blok kodu
...
else:
    blok kodu wykonywany wtedy, gdy wczesniejsze warunki zwrócą False
```

W Pythonie rozróżniamy 3 wersje polecenia `if ... else`:
1. `if`
2. `if ... else`
3. `if ... elif ... else`

**Zagnieżdżanie instrukcji warunkowych**

```
if warunek:
  # blok kodu, który się wykona, jeśli warunkek będzie spełniony (wartość True)
  if wewnętrzny warunek:
    wewnętrzny blok kodu, który się wykona, jeśli wewnętrzny oraz zewnętrzny warunkek będzie spełniony (wartość True)
```

## 2.2 Dopasowanie wzorca

Tak, struktura `match...case` (znana także jako *pattern matching*) jest stosunkowo nowszą konstrukcją w niektórych językach programowania, która służy do porównywania wyrażenia z zestawem wzorców i wykonania kodu związanego z pierwszym pasującym wzorcem. Jest to bardziej zaawansowana i elastyczna wersja instrukcji `switch` lub `case` dostępnej w wielu językach.

 Można to uznać za bardziej rozbudowaną formę instrukcji `if...elif...else`, umożliwiającą sprawdzenie wielu warunków w bardziej zorganizowany sposób.

 Przykład użycia `match...case` w Pythonie:

 W tym przykładzie, w zależności od wartości przekazanej do funkcji `opis_dnia`, `match` porównuje ją z kolejnymi wzorcami zdefiniowanymi w instrukcjach `case` i wykonuje kod powiązany z pasującym wzorcem. Wyrażenie `_` w ostatnim case działa jak **"domyślny" wzorzec**, który pasuje do wszystkiego, co nie zostało wcześniej dopasowane — podobnie jak `else` w instrukcji warunkowej.

````python
def opis_dnia(dzien):
    match dzien:
        case "poniedziałek":
            return "Początek tygodnia."
        case "piątek":
            return "Prawie weekend."
        case "sobota" | "niedziela":
            return "Weekend!"
        case _:
            return "Zwykły dzień tygodnia."

day_of_week = input("jaki dziś dzień tygodnia?")
print(opis_dnia(day_of_week))
````

**Iteracja przez zakres liczb**

Funkcja `range()` w Pythonie jest wbudowanym narzędziem służącym do generowania sekwencji liczb. Funkcja `range()` może być wywołana w kilku różnych sposobach, w zależności od potrzebnej sekwencji liczb:

* `range(stop)`
  * Generuje sekwencję liczb od `0` do `stop-1`.
  * Przyjmuje jeden argument, który określa punkt końcowy sekwencji (nie jest włączony do sekwencji).
* `range(start, stop)`
  * Generuje sekwencję liczb od `start` do `stop-1`.
  * Przyjmuje dwa argumenty: start określa punkt początkowy sekwencji, a stop punkt końcowy (nie jest włączony do sekwencji).
* `range(start, stop, step)`
  * Generuje sekwencję liczb, zaczynając od `start`, kończąc na `stop-1`, zwiększając wartość o `step` za każdym razem.
  * `step` może być również ujemny, co pozwala na generowanie sekwencji malejących.

````python
# pętla for z funkcją range
for i in range(10,1,-2):
  #print('hello')
  print(i)
````

**Iteracja przez listę**

**Iteracja przez kilka list jednocześnie**

Funkcja `zip()` jest jednym z najprostszych i najczęściej używanych sposobów na iterację przez wiele list równocześnie. Funkcja `zip()` łączy elementy z kilku iterowalnych obiektów (takich jak listy) w krotki, które można następnie przetwarzać w pętli `for`.

Jeśli listy mają różne długości, `zip()` zakończy iterację na najkrótszej liście.

````python
lista1 = [1, 2, 3, 4]
lista2 = ['a', 'b', 'c','d','e']
lista3 = [0.1, 0.2, 0.3]

for liczba, litera, liczba_zmiennoprzecinkowa in zip(lista1, lista2, lista3):
    print(liczba, litera, liczba_zmiennoprzecinkowa)
````

## 2.7 Wypełniacz

Instrukcja `pass` w Pythonie jest specjalnym operatorem, który służy jako wypełniacz lub miejsce zastępcze. Nie wykonuje żadnej operacji (jest to tzw. "no-operation" lub NOP w terminologii innych języków programowania). Jego głównym zastosowaniem jest zapewnienie syntaktycznej poprawności w miejscach, gdzie wymagane jest zdefiniowanie jakiegoś bloku kodu, ale faktycznie nie chcemy w tym miejscu niczego wykonywać.

Instrukcja `pass` jest przydatna, gdy potrzebujemy zadeklarować strukturę, ale jeszcze nie jest ona gotowa lub nie potrzebujemy jeszcze jej wypełniać konkretną logiką. **Pozwala to na utrzymanie klarowności i czytelności kodu podczas etapu planowania lub prototypowania.**

**Ważne:** Ponieważ `pass` jest ignorowany przez interpreter Pythona, jego użycie nie wpływa na działanie programu poza zapewnieniem poprawności syntaktycznej.

**Wypełniacz w instrukcjach warunkowych:** Jeśli trzeba tymczasowo zignorować określony warunek bez usuwania całego bloku `if`, można użyć `pass` jako miejsca zastępczego.
```
if warunek_jest_prawdziwy:
    pass  # TODO: Dodaj logikę później
else:
    wykonaj_inna_operacje()
```
**Wypełniacz w pętlach:** Można użyć `pass` w pętlach, które są w trakcie projektowania lub debugowania, by tymczasowo pominąć wykonanie pewnych operacji.
```
for element in kolekcja:
    pass  # Pętla nic nie robi dla każdego elementu
```

# 3.Sortowanie - listy

Funkcja `sorted()` w Pythonie jest wbudowaną funkcją, która zwraca nową posortowaną listę z elementów dowolnego iterowalnego obiektu (takiego jak lista, krotka, słownik, a nawet ciąg znaków). Jest to uniwersalne narzędzie do sortowania, które pozwala na niemal natychmiastowe uporządkowanie danych według różnych kryteriów.

Składnia:

`sorted(iterable, *, key=None, reverse=False)`

* `iterable` - obiekt iterowalny, którego elementy mają być posortowane.
* `key` - parametr opcjonalny określający funkcję jednego argumentu, która jest używana do wyodrębnienia klucza porównawczego dla każdego elementu z `iterable`. Domyślnie, jeśli `key` nie jest podane, elementy są sortowane bezpośrednio bez żadnej modyfikacji.
* `reverse` - parametr opcjonalny, jeśli ustawiony na True, powoduje, że lista jest sortowana w odwrotnej kolejności (malejąco). Domyślnie ustawiony na False, co oznacza sortowanie rosnące.

**Sortowanie listy liczb**

````python
liczby = [3, 1, 4, 1, 5, 9, 2]
print(sorted(liczby, reverse=False))
````

**Sortowanie listy słów**

````python
slowa = ['jabłko', 'pomarańcza', 'banan', 'śliwka']
print(sorted(slowa))
````

**Sortowanie w odwrotnej kolejności**

````python
liczby = [3, 1, 4, 1, 5, 9, 2]
print(sorted(liczby, reverse=True))
````

**Sortowanie z użyciem klucza**

*Przykład 1: *Sortowanie listy słów według ich długości

````python
slowa = ['jabłko', 'pomarańcza', 'banan', 'śliwka']
print(sorted(slowa, key=len, reverse=True))
````
