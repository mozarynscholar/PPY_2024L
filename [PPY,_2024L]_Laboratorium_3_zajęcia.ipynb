# 1.Generowanie liczb losowych


W Pythonie do tworzenia liczb losowych można wykorzystać moduł `random`, który oferuje szeroki zakres funkcji do generowania liczb losowych różnego typu. Moduł ten jest częścią standardowej biblioteki Pythona, więc nie wymaga instalacji dodatkowych pakietów.

**Generowanie liczby losowej z określonego zakresu:**

`random.randint(a, b)`: generuje liczbę całkowitą losowo wybraną z zakresu od `a` do `b` (włącznie z obiema tymi wartościami).

````python
import random
liczba = random.randint(1, 10)  # Generuje losową liczbę całkowitą od 1 do 10.
print(liczba)
````

**Generowanie liczby zmiennoprzecinkowej:**

* random.random(): generuje losową liczbę zmiennoprzecinkową z zakresu od 0 do 1 (włącznie z 0, ale wyłączając 1).
* random.uniform(a, b):  liczbę zmiennoprzecinkową z zakresu `(a,b)`, przy założeniu rozkładu jednostajnego (jednorodnego).

````python
liczba = random.random() # Generuje losową liczbę zmiennoprzecinkową od 0 do prawie 1.
print(liczba)

liczba2 = random.uniform(1, 10)  # Generuje losową liczbę zmiennoprzecinkową od 1 do 10.
print(liczba2)
````

**Generowanie losowych liczb z określonego rozkładu:**

Moduł random oferuje również funkcje do generowania liczb z różnych rozkładów statystycznych, np. random.gauss(mu, sigma) dla rozkładu normalnego (gaussowskiego).

````python
liczba = random.gauss(0, 1)  # Generuje liczbę z rozkładu normalnego o średniej 0 i odchyleniu standardowym 1.
print(liczba)
````

**Wybieranie losowego elementu z listy:**

`random.choice(sekwencja)`: wybranie losowego elementu z listy (lub innej sekwencji).

````python
lista = [1, 2, 3, 4, 5]
element = random.choice(lista)  # Wybiera losowy element z listy.
print(element)
````

**Mieszanie elementów listy:**

`random.shuffle(lista)`: dokonuje przemieszania elementy listy. **Funkcja ta modyfikuje oryginalną listę.**

````python
lista = [1, 2, 3, 4, 5]
print("lista przed wymieszaniem", lista)
random.shuffle(lista)  # Miesza listę w miejscu.
print("lista po wymieszaniu:", lista)
````

### 2.1.0 Mutowalność

**Mutowalność** w kontekście programowania, a szczególnie w języku Python, odnosi się do zdolności obiektu do zmiany swojego stanu lub zawartości po jego utworzeniu. W Pythonie niektóre typy danych są mutowalne (modyfikowalne), a inne są niemutowalne (niemodyfikowalne), co ma znaczący wpływ na sposób, w jaki są używane w programach.

### 2.1.1. Listy

Typ `list` w Pythonie jest jednym z wbudowanych typów danych zbiorowych (kolekcyjnych), który służy do przechowywania wielu elementów jako pojedynczą jednostkę. Lista jest uporządkowaną kolekcją, co oznacza, że **elementy w liście są przechowywane w określonej kolejności, od lewej do prawej, począwszy od indeksu 0**. Listy są zmienne (mutable), co oznacza, że można modyfikować ich zawartość po utworzeniu, dodając, usuwając lub zmieniając elementy.

**Kluczowe cechy list w Pythonie:**

* **Uporządkowane:** Kolejność elementów w liście ma znaczenie i jest zachowana.
* **Zmienne:** Można zmieniać, dodawać i usuwać elementy w liście.
* **Elastyczne:** Listy mogą przechowywać elementy różnych typów, w tym inne listy, co umożliwia tworzenie list zagnieżdżonych.
* **Dynamiczne**: Rozmiar listy może być zmieniany dynamicznie, co oznacza, że Python automatycznie zarządza alokacją pamięci dla list.

W Pythonie mamy możliwość **indeksowania listy od końca za pomocą ujemnych indeksów**. Wówczas indeksowanie zaczynamy od -1 w dół.

*Wycinanie (ang. slicing)*

Możemy wyciąć fragment listy używając **operatora do wycinania** `:`

`my_list[start:stop]`

**UWAGA:** Wycinając pierwszy podany indeks jest włączny, a ostatni rozłączny.

`y = x[0:-1:2]`

`y = x[::-1]`

Listy możemy także iterować w parach lub więcej dzięki metodzie `zip`

Metoda `zip()` w Pythonie to wbudowana funkcja, która umożliwia równoległą iterację po wielu sekwencjach, takich jak listy, krotki czy stringi. Funkcja ta "zapina" (stąd nazwa zip) elementy z podanych sekwencji, tworząc z nich krotki, gdzie każda krotka zawiera po jednym elemencie z każdej sekwencji. Długość wynikowej sekwencji generowanej przez `zip()` jest równa długości najkrótszej z sekwencji wejściowych.

Metoda `zip()` jest szczególnie przydatna w sytuacjach, gdy potrzebujemy wykonać operacje na wielu sekwencjach w sposób równoległy, jak na przykład przy porównywaniu elementów z różnych list czy tworzeniu słownika z dwóch list, gdzie jedna stanowi klucze, a druga wartości.

Przy rozpakowywaniu korzystamy z funkcji `zip(*zipped)` gdzie Operator `*` przed zmienną powoduje, że zmienna jest traktowana jako sekwencja argumentów, nie jako pojedyncza jednostka. Jest to nazywane rozpakowywaniem argumentów funkcji. Gdy używamy `*` z funkcją, taką jak `zip()`, mówimy Pythonowi, aby traktował elementy iterowalnego obiektu (np. listy, krotki) jako niezależne argumenty funkcji.

````python
languages = ['Python', 'Swift', 'C++', 'C']
programmers = ['Mark', 'Ann', 'Debbie']

for lang, programmer in zip(languages, programmers):
  print(f'{programmer} codes in {lang}')

#przykłady "zapinania" list
sekwencja1 = [1, 2, 3, 4]
sekwencja2 = ['a', 'b', 'c']
zipped = zip(sekwencja1, sekwencja2)
print(list(zipped))

#przykład rozpinania list
zipped = [(1, 'a'), (2, 'b'), (3, 'c')]
sekwencja1, sekwencja2 = zip(*zipped)
print(list(sekwencja1))
print(list(sekwencja2))
````

**Python List Comprehension**

"Python List Comprehension" to kompaktowy sposób na generowanie list w Pythonie poprzez osadzenie wyrażenia wewnątrz nawiasów kwadratowych. Pozwala to na transformację i filtrowanie kolekcji danych w jednej, zwięzłej linijce kodu, zamiast używać wielolinijkowych pętli for i instrukcji warunkowych. List comprehension jest nie tylko bardziej syntetycznie zwięzłe, ale często również efektywniejsze pod względem wydajności niż tradycyjne pętle.

Podstawowa skladnia to:

`[nowy for element in kolekcja if warunek]`

* `nowy`: wyrażenie zwracające wartość, które będzie elementem nowej listy.
* `element`: zmienna tymczasowa, która przyjmuje wartość każdego elementu z oryginalnej kolekcji (kolekcja) po kolei.
* `kolekcja`: lista, krotka, zbiór, słownik, czy inny iterowalny obiekt, z którego elementy są pobierane.
* `if warunek` to opcjonalna część, która pozwala na włączenie do nowej listy tylko tych elementów, które spełniają dany warunek.

````python
numbers = [x*x for x in range(1, 6)]
print("numbers:",numbers)
````

````python
#kod równoważny
numbers = [] #tworzy pustą listę

for x in range(1, 6):
    numbers.append(x * x)

print("numbers:", numbers)
````

````python
# Tylko parzyste liczby
parzyste = [x for x in range(10) if x % 2 == 0]
print("liczby parzyste:", parzyste)
````

### 2.1.2. Krotki

Krotka (ang. *tuple*) to uporządkowana i niemutowalna kolekcja elementów. Uporządkowanie oznacza, że elementy w krotce są przechowywane w określonej kolejności, a niemutowalność, że po jej utworzeniu nie można zmieniać jej elementów (w przeciwieństwie do list, które są mutowalne).

**Własności krotek**

* **Niemutowalność** Krotki nie mogą być zmieniane po utworzeniu. Próba zmiany, dodania lub usunięcia elementów z krotki skutkuje błędem.
* **Uporządkowanie:** Elementy w krotce mają określoną kolejność, która jest zachowana.
* **Indeksowanie:** Można uzyskać dostęp do elementów krotki za pomocą indeksów, podobnie jak w przypadku list.
* **Możliwość zawierania różnych typów danych:** Krotki mogą przechowywać elementy różnych typów, w tym inne krotki, listy, słowniki i inne.

Aby zdefiniować krotkę zawierającą elementy, musisz umieścić je wewnątrz nawiasów okrągłych, tworząc krotkę. Jeśli jest tylko jeden element należy dodać po nim przecinek. Przecinek jest konieczny, aby Python rozpoznał strukturę jako krotkę, nawet jeśli zawiera tylko jeden element.

**W Pythonie krotki są często używane do przypisywania wartości do wielu zmiennych jednocześnie.**

Krotki są podobne do list i są wykorzystywane w wielu podobnych sytuacjach.

**Ale krotki mają przewagę nad listami w kilku aspektach:**

* Korzystamy z krotek do danych o wielu typach, a list do danych o jednym typie.
* Ponieważ krotki są **niemutowalne**, ich iteracja jest szybsza niż list i ich zastosowanie sprzyja **optymalizacji kodu**.
* Krotki, które są **niemutowalne**, mogą służyć jako **klucze** w słowniku, w przeciwieństwie do list.
* Jeżeli twoje dane nie ulegną zmianie, krotki **zapewniają lepszą ochronę przed modyfikacjami**.

````python
# Definiowanie krotki
moja_krotka = (1, "Hello", 3.14,)

# Tworzenie krotki bez nawiasów
moja_krotka2 = 2, "Welcome", 10.1

# Dostęp do elementów krotki
print(moja_krotka[0])  # Wypisze 1
print(moja_krotka[1])  # Wypisze "Hello"
print(moja_krotka[2])  # wypisze 3.14
# Próba zmiany elementu krotki (spowoduje błąd)
# moja_krotka[1] = "World"  # TypeError: 'tuple' object does not support item assignment

# Przypisanie wartości z krotki do zmiennych
a, b, c = moja_krotka
print(a, type(a))  # Wypisze 1
print(b, type(b))  # Wypisze "Hello"
print(c, type(c))  # Wypisze 3.14

# Przykładowa krotka, gdzie jest tylko jeden element - lista
moja_krotka = ([100, 200, 300],) #jesli nie bedzie przecinka to nie zdefinuje tego jako krotki
print(type(moja_krotka))

# Dostęp do listy
moja_krotka[0][0]=1
print(moja_krotka[0][0])
````

### 2.1.3. Zbiory

Zbiory w Pythonie (ang. *sets*) to kolekcje, które przechowują unikalne (niepowtarzajace się) elementy w nieuporządkowany sposób. W Pythonie zbiory są zdefiniowane za pomocą nawiasów klamrowych `{}` lub za pomocą wbudowanej funkcji `set()`. **Zbiory automatycznie usuwają duplikaty elementów, pozostawiając każdy element unikalny.**

Zbiory są podobne do matematycznych zbiorów i zapewniają szybki dostęp do przechowywanych elementów oraz możliwość wykonywania operacji na zbiorach, takich jak suma, przecięcie, różnica i różnica symetryczna.

**Zbiory są szczególnie użyteczne w sytuacjach, gdy potrzebujemy przechowywać zestaw unikalnych elementów i wykonywać operacje charakterystyczne dla teorii zbiorów.**

**UWAGA**: Jeśli chcemy uzyskać dostęp do konkretnego elementu zbioru, należy przekształcić zbiór na listę lub krotkę, które wspierają indeksowanie.

Własności

* **Unikalność elementów**: W zbiorze każdy element występuje tylko raz.
* **Nieuporządkowanie**: Elementy w zbiorze nie mają określonej kolejności, a ich kolejność może różnić się przy każdym wyświetleniu.
* **Mutowalność**: Zbiory w Pythonie są mutowalne, co oznacza, że można dodawać i usuwać elementy po utworzeniu zbioru. **Niemniej jednak, same elementy przechowywane w zbiorze muszą być typu niemutowalnego (np. liczby, napisy, krotki)**.
* **Brak indeksowania**: Ze względu na ich nieuporządkowaną naturę, elementów zbioru nie można indeksować ani wywoływać przez indeks.

**UWAGA:** Zbiór może składać się z dowolnej liczby obiektów i typów mieszanych, ale **nie może zawierać obiektów mutowalnych jak listy, zbiory czy słowniki**

````python
# Utworzenie zbioru
moj_zbior = {1, 2, 3, 4, 5}

# Dodawanie elementu do zbioru
moj_zbior.add(6)
print(moj_zbior)  # Wypisze coś w stylu: {1, 2, 3, 4, 5, 6}

# Usuwanie elementu ze zbioru
moj_zbior.remove(3)
print(moj_zbior)  # Wypisze coś w stylu: {1, 2, 4, 5, 6}

# Sprawdzenie, czy element jest w zbiorze
print(2 in moj_zbior)  # Wypisze: True
print(10 in moj_zbior)  # Wypisze: False

# Operacje na zbiorach
inny_zbior = {4, 5, 6, 7, 8}
print(moj_zbior.union(inny_zbior))  # Suma zbiorów
print(moj_zbior.intersection(inny_zbior))  # Przecięcie zbiorów
````

**Tworzenie pustego zbioru**

````python
# tworzymy pusty zbiór
empty_set = set()

# tworzymy pusty słownik
empty_dictionary = { }

# sprawdzamy typ
print('Data type of empty_set:', type(empty_set))

print('Data type of empty_dictionary', type(empty_dictionary))
````

**Przykład:** znajdowanie dominanty

**Dominanta**, znana również jako **modalna**, to wartość występująca najczęściej w zbiorze danych, niezależnie od tego, czy są to dane liczbowe czy nieliczbowe (jak teksty lub kategorie). **Zbiór danych może posiadać więcej niż jedną dominantę (zjawisko multimodalności), gdy dwie lub więcej wartości występują z tą samą, najwyższą częstotliwością, lub może nie mieć żadnej dominanaty, jeśli wszystkie wartości pojawiają się równie często**. Jako jedna z miar **tendencji centralnej**, obok średniej arytmetycznej i mediany, dominanta wyróżnia się tym, że skoncentrowana jest na częstotliwości pojawiania się określonych wartości w zbiorze, a nie na ich położeniu czy sumie, co sprawia, że jest wyjątkowo użyteczna w analizie różnorodnych typów danych.

Poniższy kod działa prawidłowo, gdy najczęstszy element jest jeden, lecz, gdy np. dwa elementy listy będą powtarzały się równie często -- skrypt wyydrukuje na ekranie tylko jeden z nich.


````python
x = [4,4,2,3,5,6,7,8,8,4,7,5,4,3,2,4]
naj = max(set(x), key = x.count)
print(naj)
````

**Dodawanie i edycja obiektów w zbiorze**

* `update()`: pozwala powiększyć zbiór o elementy z innej kolekcji, np. listy. **Powtarzające się elementy zostaną usunięte.**

**Operacje na zbiorach**

* **Suma zbiorów**: Suma zbiorów A i B to wszystkie elementy w zbiorach A i B.

Możemy wykorzystać operator `|` albo metodę `union()` aby wykonać opercję sumy zbiorów.

Ważne: Operacje `A|B` oraz `A.union(B)` są równoważne A ⋃ B.

obraz.png

* **Część wspólna**: tylko elementy występujące w obu zbiorach A i B.

Możemy wykorzystać operator `&` lub metodę `intersection()` aby wykonać operację części wspólnej.

**Ważne**: Operacje `A&B` oraz `A.intersection(B)` są równoważne A ⋂ B.

obraz.png

* **Różnica zbiorów**: Różnica zbiorów A i B zawiera elementy ze zbioru A, których nie ma w zbiorze B.

Używamy operatora `-` albo metody `difference()` aby obliczyć różnicę zbiorów.

**UWAGA:** `A - B` oraz `A.difference(B)` są równoważne `A - B`.



obraz.png



* **Różnica symetryczna zbiorów**: wszystkie elementy ze zbiorów A i B poza częścią wspólną.

Możemy wykorzystać operator `^` albo metodę `symmetric_difference()` aby wykonać operachę różnicy symetrycznej zbiorów.

**Ważne**: `A^B` oraz `A.symmertric_difference(B)` są równoważne A Δ B.

````python
# pierwszy zbiór
A = {1, 3, 5}

# drugi zbiór
B = {0, 3, 2, 4}

# suma za pomocą |
print('Union using |:', A | B)

# suma za pomocą union()
print('Union using union():', A.union(B))
# część wspólna za pomocą  intersection()
print('Intersection using intersection():', A.intersection(B))
# różnica z difference()
print('Difference using difference():', A.difference(B))
# różnica symetryczna z symmetric_difference()
print('using symmetric_difference():', A.symmetric_difference(B))
````

### 2.1.4. Słowniki

Słownik (ang. *dictionary*) to nieuporządkowana kolekcja danych, w której każdy element składa się z pary klucz-wartość. Klucze są unikalne w obrębie słownika, co oznacza, że każdy klucz może pojawić się tylko raz. Wartości mogą być dowolnego typu danych i nie muszą być unikalne. Słowniki są dynamiczne, co oznacza, że można je zmieniać, dodając, usuwając lub modyfikując pary klucz-wartość. Aby zdefiniować slownik należy umieścić jego elementy (pary klucz:wartość) wewnątrz nawiasów klamrowych.

**Własności**

* **Nieuporządkowanie:** Elementy w słowniku nie mają określonej kolejności.
* **Mutowalność:** Można modyfikować słowniki po ich utworzeniu, dodając, usuwając lub zmieniając ich zawartość.
* **Indeksowane przez klucze:** Do elementów słownika dostęp uzyskuje się za pomocą kluczy, a nie indeksów liczbowych.
* **Unikalne klucze:** Każdy klucz w słowniku musi być unikalny.
* **Różnorodność typów:** Klucze zazwyczaj są napisami lub liczbami, ale mogą być dowolnym niemutowalnym typem danych; wartości mogą być dowolnego typu.

````python
# Definiowanie słownika
moj_slownik = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie']
}

# Dostęp do wartości za pomocą klucza
print(moj_slownik['imie'])  # Wypisze 'Jan'

# Dodawanie nowej pary klucz-wartość
moj_slownik['miejsce zamieszkania'] = 'Warszawa'

# Modyfikacja istniejącej wartości
moj_slownik['wiek'] = 31

# Usuwanie pary klucz-wartość
del moj_slownik['hobby']

# Wypisanie całego słownika
print(moj_slownik)
````

**Dodawanie wartości do słownika**

`nazwa_słownika[nowy_klucz] = nowa_wartość`

````python
moj_slownik = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie'],
    'miejsce zamieszkania' : 'Warszawa'
}
print("słownik przed modyfikacją:", moj_slownik)

moj_slownik['miasto']='Krakow'
print("słownik po modyfikacji:", moj_slownik)
````

Możemy dodać wiele par `klucz:wartość` z wykorzystaniem funckji `update()`

````python
moj_slownik1 = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie']
}

moj_slownik2 = {
     'Wzrost': 180.0,
     'Waga': 80.0,
     'wiek': 31
}

moj_slownik1.update(moj_slownik2)
print(moj_slownik1)
````

**Zmiana wartości w słowniku**

Również odwołujemy się do wartości przez klucz w nawiasach [].

````python
moj_slownik = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie'],
    'miasto':'Krakow'
}
print("słownik przed modyfikacją:", moj_slownik)

moj_slownik['miasto']='Warszawa'
print("słownik po modyfikacji:", moj_slownik)
````

**Dostęp do elementów w słowniku**

Aby uzyskać dostęp do wartości przypisanej do danego klucza, używamy nawiasów kwadratowych `[]`, podając klucz w nawiasach. Jeśli klucz istnieje w słowniku, zostanie zwrócona odpowiadająca mu wartość. Jeśli klucz nie istnieje, Python zgłosi wyjątek `KeyError`.

````python
moj_slownik = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie']
}
print("wartość po danym kluczu:", moj_slownik['imi'])
````

**Obsługa wyjątku `KeyError`**

Gdy próbujemy uzyskać dostęp do wartości za pomocą klucza, który nie istnieje w słowniku, Python zgłosi błąd `KeyError`. Aby tego uniknąć, można użyć metody `get()`, która zwróci `None` lub zdefiniowaną przez użytkownika domyślną wartość, jeśli klucz nie istnieje.

````python
moj_slownik = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie']
}

miasto = moj_slownik.get('miasto')
print(miasto)  # Wypisze: klucz nie istnieje
````

**Usuwanie elementów ze słownika**

Aby usunąć wybraną parę ze słownika korzystamy z instrukcji `del` i odnosimy się do `klucza`. Zostanie usunięta para `klucz:wartość`

````python
moj_slownik = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie']
}

del moj_slownik['wiek']
print(moj_slownik)
````

Korzystając z `del` możemy także usunąć cały słownik.

Chcąc skasować wszystkie elementy ze słownika korzystamy z funkcji `clear()`

````python
moj_slownik = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie']
}

moj_slownik.clear()
print(moj_slownik)

del moj_slownik
print(moj_slownik) #usuwa całą kolekcję
````

**Iteracja słownika**

Operacje na wszystkich wartościach z wykorzystaniem pętli `for `

````python
moj_slownik = {
    'imie': 'Jan',
    'nazwisko': 'Kowalski',
    'wiek': 30,
    'hobby': ['czytanie', 'rower', 'programowanie']
}

for klucz in moj_slownik.keys():
  print(f'kluczowi {klucz} przypisano wartość: {moj_slownik[klucz]}')

for wartosc in moj_slownik.values():
  print(f'wartosc {wartosc}')
````
